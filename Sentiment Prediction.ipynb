# Sentiment Analysis

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
```

```python
import pandas as pd
import seaborn as sns
```

```python
test_sentence = "$TSLA deliveries expected shortly, Tesla second-Quarter Deliveries could clear 200,000, set record. Company likely overcame chip shortage and congestion at ports. Faster, refreshed Model S plaid was introduced last month per Bloomberg"
```

```python
sid.polarity_scores(test_sentence)['compound']
```

```python
df = pd.read_csv('cleanData1.csv')
```

```python
df = df[['date', 'tweet']]
```

```python
df.head()
```

```python
current_date = df['date'][0]
polarity_sum = 0
num_tweets_in_day = 0
avg_polarity = 0
avg_polarity_list = []
dates = []

for ind in df.index:
    new_date = df['date'][ind]
    tweet = df['tweet'][ind]
    polarity = sid.polarity_scores(tweet)['compound']
    
    if current_date == new_date:
        polarity_sum += polarity
        num_tweets_in_day += 1
    elif current_date != new_date:
        avg_polarity = polarity_sum / num_tweets_in_day
        avg_polarity_list.append(avg_polarity)
        dates.append(current_date)
        
        current_date = new_date

        # reset variables
        polarity_sum = 0
        num_tweets_in_day = 0
        avg_polarity = 0
        polarity_sum += polarity    
    
    
#     print("Date:",new_date)
#     print("Polarity of the tweet:",polarity)
    
# def calculate_polarity_avg:
#     print("calculate polarity average")
```

```python
len(avg_polarity_list)
avg_polarity_list
dates
```

```python
dates.reverse()
avg_polarity_list.reverse()
polarities_df = pd.DataFrame(dates, columns = ['date'])
polarities_df['polarity'] = avg_polarity_list
```

```python
polarities_df.to_csv("AveragePolarities_Dates.csv")
```

## Time series forecasting with ARIMA
(Autoregressive Integrated Moving Average)

```python
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
print('Examples of parameter combinations for Seasonal ARIMA...')
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[2]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[3]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[4]))
```

```python
# parameter Selection for our polarities' ARIMA Time Series Model
# grid search to find the optimal set of parameters that yields the best performance for our model
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(data,
                                            order=param,
                                            seasonal_order=param_seasonal,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)

            results = mod.fit()

            print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
        except:
            continue
```

```python
# optimal option - lowest AIC value (bottom in the above cell's output)
# ARIMA(1, 1, 1)x(1, 1, 1, 12)12 - AIC:-42.30578681271336
# SARIMAX(1, 1, 1)x(1, 1, 1, 12) gives lowest AIC value

mod = sm.tsa.statespace.SARIMAX(data,
                                order=(1, 1, 1),
                                seasonal_order=(1, 1, 1, 12),
                                enforce_invertibility=False)
results = mod.fit()
print(results.summary().tables[1])
```

### Model Diagnostics

```python
results.plot_diagnostics(figsize=(16, 8))
plt.show()
plt.savefig('model diagnostics')
```

## Validating forecasts

```python
pred = results.get_prediction(start=pd.to_datetime('2021-06-16'), dynamic=False) # choose a date to start predictions
pred_ci = pred.conf_int()
ax = data['2021-05-21':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7, figsize=(14, 7))
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)
ax.set_xlabel('Date')
ax.set_ylabel('Tweet Polarities')
plt.legend()
plt.show()
```

```python
data_forecasted = pred.predicted_mean
data_truth = data['2021-06-16':]
mse = ((data_forecasted - data_truth) ** 2).mean()
print('The Mean Squared Error of our forecasts is {}'.format(round(mse, 4))) # 0.0024
print('The Root Mean Squared Error of our forecasts is {}'.format(round(np.sqrt(mse),4))) # 0.0492
```

## Producing and visualizing forecasts

```python
pred_uc = results.get_forecast(steps=100)
pred_ci = pred_uc.conf_int()
ax = data.plot(label='observed', figsize=(14, 7))
pred_uc.predicted_mean.plot(ax=ax, label='Forecast')
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.25)
ax.set_xlabel('Date')
ax.set_ylabel('Polarities')
plt.legend()
plt.show()
data.to_csv('Data')
```
